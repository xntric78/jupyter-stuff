```python
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline

imgPath = "characters/"
fileName = "3"
ext = ".jpg"

#Use PLA to convert image to greyscale. 
im = Image.open(imgPath+fileName+ext).convert('LA')
bg = Image.new("RGB", im.size, (255,255,255))
bg.paste(im)
bg.save(imgPath+fileName+".png")

img  = plt.imread(imgPath+fileName+".png")
imgplot = plt.imshow(img)
```
